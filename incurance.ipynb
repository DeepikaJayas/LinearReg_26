```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv("insurance_data.csv")
print(df.head())
```

```python
X = df[['age']]             
y = df['bought_insurance']
```

```python
model = LogisticRegression()
model.fit(X, y)
```

```python
print(model.predict([[30]]))   
print(model.predict([[45]]))

print(model.predict_proba([[30]]))  
print(model.predict_proba([[45]]))
```

```python
age_range = np.linspace(df.age.min(), df.age.max(), 100).reshape(-1,1)
probabilities = model.predict_proba(age_range)[:,1]
```

```python
plt.scatter(df.age, df.bought_insurance, color='red', marker='1')

plt.plot(age_range, probabilities)

plt.xlabel("Age")
plt.ylabel("Probability of Buying Insurance")
plt.title("Logistic Regression Curve")

plt.show()
```

```python
print(model.coef_)
print(model.intercept_)
```
